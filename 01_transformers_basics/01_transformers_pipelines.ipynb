```python
from transformers import pipeline
```

**Sentiment Analysis**

```python
classifier = pipeline('sentiment-analysis')
classifier("Hurry! i won the match haha")
```

```python
# passing saveral sentences
classifier(
    ['Hey, My name is Muqeet.', 'yes i found it.', 'I hate this too much.']
)
```

**Zero-shot Classification**

```python
classifier = pipeline('zero-shot-classification')
classifier(
    "I'm learning LLM",
    candidate_labels=['education','entertainment','politics']
)
```

```python
classifier(
    "cat is playing with a ball",
    candidate_labels=['education','entertainment','politics']
)
```

```python
classifier(
    "He sells his laptop",
    candidate_labels=['entertainment','business','politics','education']
)
```
